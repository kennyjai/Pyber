```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# convert to dataframe
city_df = pd.DataFrame(city_df)
ride_df = pd.DataFrame(ride_df)

# Combine the data into a single dataset
merge_df = pd.merge(ride_df, city_df, on="city")

# Display the data table for preview
merge_df.set_index('city')
merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build a Bubble Plot for 4 key variables: avg fare, total rides, total drivers, city type

# Obtaining x and y coordinates for urban (city type)
urban_df = merge_df.loc[merge_df['type'] == 'Urban']

# Group by city so we can get better calculations
urban_df = urban_df.groupby('city')

# Calculate average fare ($) per city (y coordinate)
urban_fare = urban_df['fare'].mean()

# Calculate total number of rides per city (x coordinate)
urban_rides = urban_df['ride_id'].count()

# Calculate total number of drivers per city (sizing for circle size to correlate with driver count per city)
urban_drivers = urban_df['driver_count'].mean()

# Repeat calculations for x and y coordinates for Suburban and Rural
# Obtaining x and y coordinates for Suburban
suburban_df = merge_df.loc[merge_df['type'] == 'Suburban']
suburban_df = suburban_df.groupby('city')

suburban_fare = suburban_df['fare'].mean()
suburban_rides = suburban_df['ride_id'].count()
suburban_drivers = suburban_df['driver_count'].mean()

# Obtaining x and y coordinates for Rural
rural_df = merge_df.loc[merge_df['type'] == 'Rural']
rural_df = rural_df.groupby('city')

rural_fare = rural_df['fare'].mean()
rural_rides = rural_df['ride_id'].count()
rural_drivers = rural_df['driver_count'].mean()

# Color labeling per city type 
colors = ["gold","lightskyblue","lightcoral"]

# Build the scatter plots for each city types

urban = plt.scatter(urban_rides, urban_fare, marker="o", facecolors=colors[2], 
                     edgecolors='black',linewidths=1,s=urban_drivers*10,alpha=0.8,label='Urban')

suburban = plt.scatter(suburban_rides, suburban_fare, marker="o", facecolors=colors[1], edgecolors='black',
            linewidths=1.2,s=suburban_drivers*10,alpha=0.8,label='Suburban')

rural = plt.scatter(rural_rides, rural_fare, marker="o", facecolors=colors[0], edgecolors='black',
                    linewidths=1.2,s=rural_drivers*10,alpha=0.8,label='Rural')

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)",fontweight='bold')
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare(s)")
plt.grid()

# Create a legend at best location
plt.legend(title="City Types",loc='best', markerscale=.5)

# Incorporate a text label regarding circle size
plt.figtext(.93,.5,"Note: Circle size correlates with driver count per city",wrap=True)

# Save Figure under Images folder named bubble plot
plt.savefig("Images/bubble_plot.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Create a group based on the city types in the 'type' column
type_group = merge_df.groupby('type')

# Calculate the sum of all the fares that appear in each city type
total_fares = type_group['fare'].sum()

# Calculate the percentage by using (fares per city / sum of total fares * 100) 
total_fares = (total_fares / total_fares.sum()) * 100

# Create data for pie chart visuals
city_type = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)

# Build Pie Chart
plt.title("% of Total Fares by City Type",fontweight='bold')
plt.pie(total_fares,explode=explode,labels=city_type,colors=colors,autopct="%1.1f%%",
        shadow=True,startangle=160)

# Save Figure
plt.savefig("Images/total_fares_by_city_type.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Use previous city_type_group

# Calculate the total unique rides by each city type
total_rides = type_group['ride_id'].count()

# Calculate the percentage by using (total rides per city / sum of total rides * 100) 
total_rides = (total_rides / total_rides.sum()) * 100

# Build Pie Chart
plt.title("% of Total Rides by City Type",fontweight='bold')
plt.pie(total_rides,explode=explode,labels=city_type,colors=colors,autopct="%1.1f%%",
        shadow=True,startangle=160)

# Save Figure
plt.savefig("Images/total_rides_by_city_type.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Create a new group to group by both city type and city because we're looking for total driver
city_type_group = merge_df.groupby(['city','type'])

# Calculate the driver percentages in each city type by using the mean
# We're using the mean because the driver count doesn't change per city
total_drivers = city_type_group['driver_count'].mean()

# Group by type so we can look at total drivers per city type
total_drivers = total_drivers.groupby('type')

# Calculate the total drivers percentage by getting the sum of drivers per city type / overall sum
total_drivers = (total_drivers.sum() / (total_drivers.sum()).sum()) * 100

# Build Pie Chart
plt.title("% of Total Drivers by City Type",fontweight='bold')
plt.pie(total_drivers,explode=explode,labels=city_type,colors=colors,autopct="%1.1f%%",
        shadow=True,startangle=160)

# Save Figure
plt.savefig("Images/total_drivers_by_city_type.png")
```

```python
# Show Figure
plt.show()
```

# Analysis

We can see from the pie charts and scatter plot that there are significantly more drivers in urban areas and they account for most of the rides and fares overall. However, if you look at Rural Drivers, although they only account for 2.6% of the total amount of drivers by city type, they account for 6.8% of the total fares and get a higher average fare. 

Another trend that we can see is that Suburban drivers similar to rural drivers. They account for 16.5% of the total drivers and earn 30.5% of the total fares. But because they account for more of the total rides, their average fare earned isn't as high compared to rural drivers. 

When we analyze urban drivers, they accout for the largest portion of total drivers, total rides, and total fares. But because they account for most of it, their average fare is lower compared to suburban and rural city types. We can assume this is due to larger amount of drivers which would affect supply and demand. Since there is more supply of drivers, the demand is less and therefore fares are less. Same can be said about rural drivers. Because there are so little rural drivers, demand is a lot more and fares are higher.
